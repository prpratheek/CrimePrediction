```python
import pandas as pd 
from sklearn.ensemble import RandomForestRegressor
data1=pd.read_csv('Data.csv')
data1.head(15)
```

```python
import numpy as np
X=np.array(data1.drop(['Class'],1))
y=np.array(data1['Class'])
```

```python
rfg=RandomForestRegressor()
rfg.fit(X,y)
```

```python
#Testing
a=np.array([[12.991776,80.261698,3,2,0,0,2],[12.929738,80.217066,1,2,0,0,2]])
print(rfg.predict(a))
```
